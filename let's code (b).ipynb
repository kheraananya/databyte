```python
import csv

with open('/Users/kheraji/Downloads/data_csv.csv', mode = "r") as infile:
     reader = csv.reader(infile)
     with open('/Users/kheraji/Downloads/data_csv_new.csv',mode = 'w') as outfile:
       writer = csv.writer(outfile)
       mydict = {rows[2]:rows[1] for rows in reader }
```

```python
mydict
```

```python
def code(string):
    i = 0
    for key in mydict:
        if(key == string):
            return i
        else:
            i += 1
```

```python
Code_1 = input('Country code in capital 1:')
Code_2 = input('Country code in capital 2:')
for key in mydict:
    if(code(key)>code(Code_1) and code(key)<code(Code_2)):
        print(mydict.get(key))
```
