```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
from matplotlib import style
%matplotlib inline
```

```python
df = pd.read_csv("/Users/kheraji/Downloads/company_sales.csv")
df
```

```python
style.use('ggplot')
plt.figure(figsize=(8,6))
plt.plot(df['month_number'], df['facecream'],  label = 'Face cream Sales Data',  color='green', linewidth=3)
plt.plot(df['month_number'], df['facewash'],   label = ' facewash Sales Data',   color='red',   linewidth=7)
plt.plot(df['month_number'], df['toothpaste'], label = 'ToothPaste Sales Data',  color='blue',  linewidth=3)
plt.plot(df['month_number'], df['bathingsoap'],label = 'bathingsoap Sales Data', color='brown', linewidth=3)
plt.plot(df['month_number'], df['shampoo'],    label = 'shampoo Sales Data',     color='black', linewidth=3)
plt.plot(df['month_number'], df['moisturizer'],label = 'moisturizer Sales Data', color='white', linewidth=3)
plt.legend(loc ='upper left')
plt.xlabel('Month Number')
plt.ylabel('Sales units')
plt.title ('Product Sales Data')
plt.xticks(df['month_number'])
plt.yticks([500,2500,4500,6500,8500,10500,12500,14500])
plt.show()
```
